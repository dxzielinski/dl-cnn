## Comparison of loss functions

```python
from mlflow.tracking import MlflowClient
import matplotlib.pyplot as plt


client = MlflowClient()
run_id = "9564c66c96294b55af87c7f98b8a9d01"
metrics = client.get_metric_history(run_id, "train_loss_epoch")
steps = [m.step for m in metrics]
values = [m.value for m in metrics]

plt.plot(steps, values, label='Train Loss')
plt.xlabel("Step")
plt.ylabel("Loss")
plt.title("Train Loss from MLflow")
plt.legend()
plt.show()
```
